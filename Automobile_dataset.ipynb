```python
#Euclidean distance between symboling and highway-mpg
temp=df[['symboling','highway-mpg']].dropna()
s1=temp.iloc[0]['symboling']
highway_mpg1=temp.iloc[0]['highway-mpg']
s2=temp.iloc[2]['symboling']
highway_mpg2=temp.iloc[2]['highway-mpg']
distance=np.sqrt((s1-s2)**2+(highway_mpg1-highway_mpg2)**2)
print("\nEuclidean Distance")
print(distance)
```

```python
#Manhattan Distance between curb-weight and engine size
temp=df[['curb-weight','engine-size']].dropna()
manhattan_distances = []
for i in range(len(temp)):
    for j in range(i + 1, len(temp)):
        curb_weight1 = temp.iloc[i]['curb-weight']
        engine_size1 = temp.iloc[i]['engine-size']
        curb_weight2 = temp.iloc[j]['curb-weight']
        engine_size2 = temp.iloc[j]['engine-size']


        distance = abs(curb_weight1 - curb_weight2) + abs(engine_size1 - engine_size2)


        manhattan_distances.append((i, j, distance))
for row1, row2, distance in manhattan_distances:
    print(f"Distance between row {row1} and row {row2}: {distance}")
```

```python
#calculating mink0wshi Distance between symboling and highway-mpg
minkowshi=((abs(s1-s2)**3)+(abs(highway_mpg1-highway_mpg2)**3))**(1/3)
print("\nMinkowshi Distance")
print(minkowshi)

```

```python
#cosine Dissimilarity
A=np.array([s1,highway_mpg1])
B=np.array([s2,highway_mpg2])
cosine_similarity=np.dot(A,B)/(np.linalg.norm(A)*np.linalg.norm(B))
cosine_dissimilarity=1-cosine_similarity
print("\nCosine Dissimilarity")
print(cosine_dissimilarity)
```
