```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("/content/Automobile_data.csv")
df
```

```python
print("printing data types")
print(df.dtypes)
```

```python
print("printing column names")
print(df.columns)
```

```python
sns.countplot(x='symboling',data=df)
plt.title("countplot for symboling")
plt.show()
```

```python
sns.histplot(x='height',data=df,kde=True)
plt.title("Height Distribution")
plt.show()
```

```python
sns.heatmap(df.corr(numeric_only=True),annot=True,cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

```python
sns.histplot(df["engine-size"], bins=20, kde=True)
plt.title("Engine Size Distribution")
plt.show()
```

```python
sns.histplot(df["wheel-base"], bins=20, kde=True)
plt.title("Wheel Base Distribution")
plt.show()
```

```python
sns.histplot(df["length"], bins=20, kde=True)
plt.title("Length Distribution")
plt.show()
```

```python
sns.histplot(df["width"], bins=20, kde=True)
plt.title("Width Distribution")
plt.show()
```

```python
sns.histplot(df["curb-weight"], bins=20, kde=True)
plt.title("Curb Weight Distribution")
plt.show()
```

```python
sns.histplot(df["city-mpg"], bins=20, kde=True)
plt.title("city-mpg Distribution")
plt.show()
```

```python
sns.histplot(df["highway-mpg"],bins=20,kde=True)
plt.title("Highway MPG Distribution")
plt.show()
```

```python
sns.histplot(df["compression-ratio"],bins=20,kde=True)
plt.title("Compression Ratio Distribution")
plt.show()
```

```python
sns.countplot(x="body-style", data=df)
plt.xticks(rotation=45)
plt.title("Body Style Distribution")
plt.show()
```

```python
sns.countplot(x="fuel-type", data=df)
plt.xticks(rotation=45)
plt.title("Fuel-type Distribution")
plt.show()
```

```python
sns.countplot(x="make", data=df)
plt.xticks(rotation=50)
plt.title("Make Distribution")
plt.show()
```

```python
sns.boxplot(x="body-style", y="engine-size", data=df)
plt.title("Body Style vs Engine Size")
plt.show()
```

```python
sns.scatterplot(x="engine-size", y="price", data=df)
plt.title("Engine Size vs Price")
plt.show()
```

```python
sns.lineplot(x="engine-size", y="price", data=df)
plt.title("Engine Size vs Price")
plt.show()
```

```python
sns.pairplot(df, hue="body-style")
plt.title("Pair Plot of Body Style")
plt.show()
```

```python
sns.violinplot(x="body-style", y="price", data=df)
plt.title("Body Style vs Price")
plt.show()

```
